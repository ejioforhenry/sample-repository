```python
# Histogram for Age_grroup
plt.hist('Age_Group', data=data, bins=4)
plt.title('Age_Group_Distribution')
plt.xlabel('Segment')
plt.ylabel('values')
plt.show()
```

```python
data['Age_Group'].value_counts().plot(kind='bar')
plt.title('Age_Group_Distribution')
plt.xlabel('Segment')
plt.ylabel('values')
plt.show()
```
